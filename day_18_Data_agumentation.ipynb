```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Flatten, MaxPooling2D, Conv2D
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img("D:/my picture/IMG20231126114417.jpg", target_size = (3000, 4000))
```

```python
# Data Agumentation
datagen = ImageDataGenerator(
    rotation_range = 40,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    brightness_range = (0.5, 1.5),
    width_shift_range = 0.2,
    height_shift_range = 0.2
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
type(img)
```

```python
input_batch = img.reshape(1, 3000, 4000, 3)
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size = 1, save_to_dir="C:/Users/Suvendu Khuntia\Desktop/deep-learning-implementations"):

  i = i+1

  if i == 10:
    break
```
