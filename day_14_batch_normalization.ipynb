```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Suvendu Khuntia/Downloads/ushape (1).csv")
```

```python
df.head()
```

```python
plt.scatter(df['X'], df['Y'], c = df['class'])
```

```python
x = df.drop(columns='class')
y = df['class']
```

```python
import tensorflow
from tensorflow import keras
from keras.models import  Sequential
from keras.layers import Dense, BatchNormalization
```

```python
model = Sequential()
```

```python
model.add(Dense(2, activation='relu', input_dim = 2))
model.add(Dense(2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=100, validation_split=0.2, verbose=2)
```

```python
plt.plot(history.history['accuracy'], label = 'train')
plt.plot(history.history['val_accuracy'], label = 'test')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

# After adding batch normalization

```python
model_1 = Sequential()
```

```python
model_1.add(Dense(2, activation='relu', input_dim = 2))
model_1.add(BatchNormalization())
model_1.add(Dense(2, activation='relu'))
model_1.add(BatchNormalization())
model_1.add(Dense(1, activation='sigmoid'))
```

```python
model_1.summary()
```

```python
model_1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history_1 = model_1.fit(x, y, validation_split=0.2,epochs=100,  verbose=2)
```

```python
plt.plot(history_1.history['accuracy'], label='Training Accuracy')
plt.plot(history_1.history['val_accuracy'], label = 'val_data')
plt.legend()
```
