```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
# define AND data
and_x = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

and_y = np.array([0,0,0,1])

# define OR data
or_x = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

or_y = np.array([0,1,1,1])


# define XOR data
xor_x = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

xor_y = np.array([0,1,1,0])
```

```python
# train a percepton on each
p_and = Perceptron()
p_and.fit(and_x, and_y)

p_or = Perceptron()
p_or.fit(or_x, or_y)

p_xor = Perceptron()
p_xor.fit(xor_x, xor_y)



# plotting setup
fig, axs = plt.subplots(1, 3, figsize=(15,5))

datasets = [
    (and_x, and_y, p_and, "AND Data"),
    (or_x, or_y, p_or, "OR Data"),
    (xor_x, xor_y, p_xor, "XOR Data"),
]

for ax, (X, y, model, title) in zip(axs, datasets):
    # plot the data points
    ax.scatter(X[y==0][:,0], X[y==0][:,1], marker='o', color='red', label='class 0')
    ax.scatter(X[y==1][:,0], X[y==1][:,1], marker='x', color='blue', label='class 1')

    # draw decision boundary
    xx, yy = np.meshgrid(np.linspace(-0.2,1.2,100), np.linspace(-0.2,1.2,100))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.2, cmap='bwr')

    ax.set_title(title)
    ax.set_xlim(-0.2,1.2)
    ax.set_ylim(-0.2,1.2)
    ax.legend()

plt.tight_layout()
plt.show()
```
